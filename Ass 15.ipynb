```python
1. Write a Python program to check if the given number is a Disarium Number?
Ans)
import math
def check(n) :
    count_digits = len(str(n))
    sum = 0
    x = n
    while (x!=0) :
 
        # Get the rightmost digit
        r = x % 10
        sum = (int) (sum + math.pow(r, count_digits))
        count_digits = count_digits - 1
        x = x//10
    if sum == n :
        return 1
    else :
        return 0
n = 32514
if (check(n) == 1) :
        print ("Disarium Number")
else :
    print ("Not a Disarium Number")
```

```python
3. Write a Python program to check if the given number is Happy Number?
Ans)
def numSquareSum(n):
    squareSum = 0;
    while(n):
        squareSum += (n % 10) * (n % 10);
        n = int(n / 10);
    return squareSum;
 

def isHappynumber(n):
    slow = n;
    fast = n;
    while(True):
        slow = numSquareSum(slow);
        fast = numSquareSum(numSquareSum(fast));
        if(slow != fast):
            continue;
        else:
            break;
    return (slow == 1);
 

n = 14;
if (isHappynumber(n)):
    print(n , "is a Happy number");
else:
    print(n , "is not a Happy number");
```

```python
4. Write a Python program to print all happy numbers between 1 and 100?
Ans)
import math
def digitsSquareSum(Number):
    Sum = rem = 0
    while Number > 0:
        rem = Number % 10
        Sum = Sum + math.pow(rem, 2)
        Number = Number // 10
    return Sum

minHpy = int(input("Enter the Minimum Happy Number = "))
maxHpy = int(input("Enter the Maximum Happy Number = "))

print("\nThe List of Happy Numbers from {0} and {1}".format(minHpy, maxHpy)) 
for i in range(minHpy, maxHpy + 1):
    Temp = i

    while Temp != 1 and Temp != 4:
        Temp = digitsSquareSum(Temp)

    if Temp == 1:
        print(i, end  = '  ')
```
